```python
#Mapping --> to map arabic to its corrosponding english coubterpart so if we call one randomly so other comes automatically
#it has 2 parts --> firt is key and sond is its value 
#hashing is used to store those value o its keys 

latin_numbers_dict={"0":"٠","1":"١","2":"٢","3":"٣","4":"٤","5":"٥","6":"٦","7":"٧","8":"٨","9":"٩"}
arabic_numbers_dict={"٠":"0","١":"1","٢":"2","٣":"3","٤":"4","٥":"5","٦":"6","٧":"7","٨":"8","٩":"9"}

#listing 
letters = ["ا", "ب", "ح", "د", "ر", "س", "ص", "ط", "ع", "ق", "ك", "ل", "م", "ن", "ه", "و" ,"ى"] 
# Lletters = {"ا" ,"":"ب", "":"ح" ,"":"د" ,"":"ر","" :"س","" :"ص","" :"ط","":"ع","" :"ق","" :"ك","" :"ل","" :"م","" :"ن","":"ه","" :"و","a" ,"ى":"v"}
arabic_letters_dict = {"ا":"A","ب":"B","ح":"J","د":"D","ر":"R","س":"S","ص":"X","ط":"T","ع":"E","ق":"G","ك":"K","ل":"L","م":"Z","ن":"N","ه":"H","و":"U","ى":"V"}
special_letters = ''
numbers = '٠١٢٣٤٥٦٧٨٩'

#initializing variables 

latin_digits = []
arabic_rand_numb = []
alphabets = []
arabic_rand_letters = []

#loops are run to have random 7 alphanumeircs 

import random

#random digits = runs for 4 times 
#append 4 digits chosen by numbers variable randomly then append into arabic_rand_numb
#then using 4 values index by varaible 1 it will see first from mapping and then select thier english counterpart and append them in variable latin_digits
#the same processs will be done for other loop of 3 as well for alphabets 

for i in range(4):
    arabic_rand_numb.append(random.choice(numbers))
    latin_digits.append((arabic_numbers_dict[arabic_rand_numb[i]]))

for i in range(3):
    arabic_rand_letters.append(random.choice(letters))
    alphabets.append((arabic_letters_dict[arabic_rand_letters[i]]))

#printing those all randomly generted alphanumeircs 

print(arabic_rand_numb)
print(latin_digits)
print(arabic_rand_letters)
print(alphabets)

#importing image and drawing libraries 
#save the plate template with the name of template and with the format of jpg in the same directory where you have kept this code
from PIL import Image, ImageFont, ImageDraw
im = Image.open("template.jpg")
d = ImageDraw.Draw(im)

#you can change the font as well but remember whatvever font you choose, download that and save it in same directory where code is 
#and replace font name in below line as well 
#draw arabic numbers

font = ImageFont.truetype("arial.ttf",150)
arabic_rand_numb_string = ("".join(map(str, arabic_rand_numb)))
d.text((30,30),arabic_rand_numb_string,fill=("black"),font =font)

#draw latin numbers 

font = ImageFont.truetype("arial.ttf",140)
latin_digits_string = ("".join(map(str, latin_digits)))
d.text((30,170),latin_digits_string,fill=("black"),font =font)

#draw arabic alphabets

font = ImageFont.truetype("arial.ttf",80)
arabic_rand_letters_string = (''.join(map(str, arabic_rand_letters)))

#logic to put arabic string in center

size=font.getsize(arabic_rand_letters_string)
print(size)
cal_size=460-size[0]/2
print(cal_size)

d.text((cal_size,50),arabic_rand_letters_string ,fill=("black"),font =font)


#drawing alphabtes 

font = ImageFont.truetype("arial.ttf",90)
alphabets_string = (''.join(map(str, alphabets)))
d.text((375,200),alphabets_string,fill=("black"),font =font)

#to save synthetic plate 

im.save("drawn_grid.png")



#to shown genertaed image on console 

im = Image.open('drawn_grid.png')
display(im)
```
